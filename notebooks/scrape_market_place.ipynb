```python
# Import necessary libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Set up base url
base_url = f"https://www.facebook.com/marketplace/"
```

```python
# Set up search parameters
days_listed = 7
city = "sanfrancisco"
query = "room"
```

```python
#Set up full url
url = f"{base_url}{city}/search?daysSinceListed={days_listed}&query={query}&exact=false"
```

```python
# Visit the website
browser.visit(url)
```

```python
if browser.is_element_present_by_css('div[aria-label="Close"]', wait_time=10):
  browser.find_by_css('div[aria-label="Close"]').first.click()
```

```python
# Scroll down to load more results

# number of times to scroll the page
scroll_count = 2

# delay (in seconds) between each scroll
scroll_delay = 3

# Loop to perform scrolling
for _ in range(scroll_count):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_delay)
```

```python
# Parse the HTML
html = browser.html

# Create a BeautifulSoup object from the scraped HTML
market_soup = soup(html, 'html.parser')
# Check if HTML was scraped correctly
market_soup
```

```python
# End the automated browsing session
browser.quit()
```
